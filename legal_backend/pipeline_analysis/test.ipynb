```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

```

```python
#load csv files and load possible laws
df = pd.read_csv('data/2013-2016_Nalaxone_Overdose_Prevention_Laws.csv')
df_sub = df.copy(deep=True)
```

```python
conds = []
conds.append("df_sub['Percent-of-Population-Under-15-(2013-2017)']>0.216")
# conds.append("df_sub['Number-of-Mental-Health-Facilities-(As-of-March-2023)']==0")
for cond in conds:
    df_sub = df_sub.loc[eval(cond)].reset_index(drop=True)
```

```python
data = dict(Counter(df_sub['state_name']))
```

```python
# Extract keys and values from the dictionary
keys = data.keys()
values = data.values()
plt.figure(figsize=(10, 6))
# Create a bar chart
plt.bar(keys, values)
plt.xlabel('State')
plt.ylabel('Quantity')
plt.xticks(rotation=90)
plt.title('Number of Counties with large U15 population by state')
plt.savefig('res/U15_state.png')
plt.show()
```
